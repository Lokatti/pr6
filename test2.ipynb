```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('vgsales_2.csv')
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

OrdEnc = OrdinalEncoder()
df[['Name','Platform','Publisher','NA_Sales']] = OrdEnc.fit_transform(df[['Name','Platform','Publisher','NA_Sales']])
df
```

```python
from sklearn.preprocessing import LabelEncoder

LabEnc = LabelEncoder()
LabEnc.fit(df['Rank'])
df['Rank'] = LabEnc.fit_transform(df['qwe'])
df
```
